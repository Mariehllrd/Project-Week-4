## Importing librairies

```python
import numpy as np
import pandas as pd
import fuzzy_pandas as fpd
```

## Importing datasets

```python
survey = pd.read_excel('./datasets/survey_clean.xlsx')
survey.head()
```

```python
immu = pd.read_csv('./datasets/immu_clean.csv')
immu.head()
```

```python
wiki = pd.read_csv('./datasets/wiki_clean.csv')
wiki.head()
```

```python
survey['Country'] = survey['Country'].str.replace('The Gambia','Gambia')
```

```python
immu['Country Name'] = immu['Country Name'].str.replace("Cote d'Ivoire",'Ivory Coast').str.replace(
    'Kyrgyz Republic','Kyrgyzstan').str.replace(
    'Lao PDR','Laos').str.replace(
    'Russian Federation','Russia').str.replace(
    'Slovak Republic','Slovakia')
```

## Merging datasets

```python
data = fpd.fuzzy_merge(survey, immu, left_on='Country', right_on='Country Name', ignore_case=True)
data = data.drop('Unnamed: 0', axis = 1)
```

```python
data = data.drop(['Country Name'], axis = 1)
```

```python
data
```

```python
data = pd.merge(data, wiki, left_on = 'Country', right_on = 'Countries', how = 'outer')
```

```python
data = data.drop(['Countries'], axis = 1)
```

```python
data = data.drop(['Unnamed: 0'], axis = 1)
```

```python
data = data[data['Country'].notna()]
```

```python
data.head()
```

```python
# Saving dataset
data.to_csv('./datasets/dataset_clean.csv')
```
